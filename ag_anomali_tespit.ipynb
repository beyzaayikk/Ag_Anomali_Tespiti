```python
%pip install missingno

import numpy as np
import pandas as pd
import seaborn as sns
import missingno as msno
sns.set(style='darkgrid')
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('dataset\Monday-WorkingHours.pcap_ISCX.csv')
data2=pd.read_csv('dataset\Tuesday-WorkingHours.pcap_ISCX.csv')
data3=pd.read_csv('dataset\Wednesday-workingHours.pcap_ISCX.csv')
data4=pd.read_csv('dataset\Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv',encoding='latin1')
data5=pd.read_csv('dataset\Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv')
data6=pd.read_csv('dataset\Friday-WorkingHours-Morning.pcap_ISCX.csv')
data7=pd.read_csv('dataset\Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
data8=pd.read_csv('dataset\Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
```

```python
data_list = [data1, data2, data3, data4, data5, data6, data7, data8]

print('Veri boyutu: ')
for i, data in enumerate(data_list, start = 1):
  rows, cols = data.shape
  print(f'Data{i} -> {rows} satır, {cols} sütun')
```

```python
data = pd.concat(data_list)
rows, cols = data.shape

print('Yeni boyut:')
print(f'Satırların sayısı: {rows}')
print(f'sütunların sayısı: {cols}')
print(f'Toplam hücre sayısı: {rows * cols}')
```

DATA CLEANING

```python
col_names={col:col.strip() for col in data.columns}
data.rename(columns=col_names,inplace=True)
```

```python
data.drop(['Flow ID','Source IP','Destination IP','Timestamp'],axis=1,inplace=True)
```

```python
duplicated_rows = data.duplicated().sum()
print(f'Tekrarlanan satır sayısı: {duplicated_rows}')
```

```python
data.drop_duplicates(inplace=True)
data.shape
```

```python
print(data.isnull().sum().sum())
```

```python
numeric_cols = data.select_dtypes(include = np.number).columns
inf_count=np.isinf(data[numeric_cols]).sum()
print(inf_count[inf_count > 0])
```

```python
print(f'İlk eksik değerlerin sayısı: {data.isnull().sum().sum()}')
data.replace([np.inf, -np.inf], np.nan, inplace=True)
print(f'Değişimden sonra eksik değerlerin sayısı: {data.isnull().sum().sum()}')
```

```python
numeric_cols = data.select_dtypes(include = np.number).columns
inf_count = np.isinf(data[numeric_cols]).sum()

# Check if there are any infinite values
if inf_count[inf_count > 0].empty:
	print("No infinite values found.")
else:
	print(inf_count[inf_count > 0])
```

```python
missing=data.isnull().sum()
mis_per=(missing/len(data))*100
mis_table=pd.concat([missing,mis_per.round(2)],axis=1)
mis_table=mis_table.rename(columns={0:'Eksik Değerler',1:'Eksik Değerlerin Yüzdesi'})
print(mis_table.loc[mis_per>0])
```

```python

sns.set_palette('pastel')
colors = sns.color_palette()

missing_vals = [col for col in data.columns if data[col].isna().any()]

fig, ax = plt.subplots(figsize = (2, 6))
msno.bar(data[missing_vals], ax = ax, fontsize = 12, color = colors)
ax.set_xlabel('Features', fontsize = 12)
ax.set_ylabel('Non-Null Value Count', fontsize = 12)
ax.set_title('Missing Value Chart', fontsize = 12)
plt.show()
```
